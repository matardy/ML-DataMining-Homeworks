(f) Create a new data set by imputing the mode in all the columns which have integer type values.

```r
# Identify columns with double type values
double_columns <- sapply(carIns, is.double)

# Impute missing values with the mean in double type columns
carInsurance_imputed_mean <- carIns
carInsurance_imputed_mean[, double_columns] <- lapply(carInsurance_imputed_mean[, double_columns], function(x) {
  if (any(is.na(x))) {
    x[is.na(x)] <- mean(x, na.rm = TRUE)
  }
  x
})
```

(g) Create a new data set by imputing the most frequent value to the column ”nDoors”.

```r
# Function to calculate mode
get_mode <- function(x) {
  unique_vals <- unique(x)
  unique_vals[which.max(tabulate(match(x, unique_vals)))]
}
```

```r
# Identify columns with integer type values
integer_columns <- sapply(carIns, is.integer)

# Impute missing values with the mode in integer type columns
carInsurance_imputed_mode <- carIns
carInsurance_imputed_mode[, integer_columns] <- lapply(carInsurance_imputed_mode[, integer_columns], function(x) {
  if (any(is.na(x))) {
    x[is.na(x)] <- get_mode(x)
  }
  x
})
```

(h) Combine the three last imputations to obtain a final dataset. Are there any duplicated cases?

```r
# Impute missing values in the "nDoors" column with the most frequent value
carInsurance_imputed_nDoors <- carIns
carInsurance_imputed_nDoors$nDoors[is.na(carInsurance_imputed_nDoors$nDoors)] <- get_mode(carInsurance_imputed_nDoors$nDoors)

# Function to calculate mode
get_mode <- function(x) {
  freq <- table(x)
  mode_val <- as.numeric(names(freq))[freq == max(freq)]
  mode_val
}
```

```r
carInsurance_imputed_nDoors
```

```r
# Combine the three imputed datasets into a final dataset
final_dataset <- bind_rows(carInsurance_imputed_mean, carInsurance_imputed_mode, carInsurance_imputed_nDoors)

# Check for duplicated cases
duplicated_cases <- final_dataset[duplicated(final_dataset), ]

# Print the duplicated cases
print(duplicated_cases)
```

## Data Pre-Processing

(a) Apply range-based normalization and z-score normalization to the price attribute.

```r
library(dplyr)

# Normalize the "price" column
carInsurance_transformed <- carIns %>%
  mutate(price = scale(price))
carInsurance_transformed
```

(b) Discretize it into 4 equal-frequency ranges an into 4 equal-width ranges.

```r
# Discretize the "price" column into 4 equal-frequency ranges (bins)
carInsurance_binned <- carIns %>%
  mutate(price_bin = cut(price, breaks = 4, labels = FALSE))

```
